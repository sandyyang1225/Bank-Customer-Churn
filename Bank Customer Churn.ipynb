## Overview
Bank customer churn refers to the situation where customers discontinue their association with a specific bank, either by closing their accounts or discontinuing their services. The analysis and understanding of customer churn hold significant importance for banks and financial institutions as it enables them to maintain their competitiveness, uphold customer satisfaction, and foster business growth. By conducting such analysis, banks can make well-informed decisions, enhance their products and services, and cultivate lasting and stronger relationships with their customers. This project aims to investigate and analyze various factors contributing to bank customer churn.

## About Dataset<br>
- __RowNumber__ : corresponds to the record (row) number and has no effect on the output.<br>
- __CustomerId__ : contains random values and has no effect on customer leaving the bank.<br>
- __Surname__ : the surname of a customer has no impact on their decision to leave the bank.<br>
- __CreditScore__ : can have an effect on customer churn, since a customer with a higher credit score is less likely to leave the bank.<br>
- __Geography__ : a customer’s location can affect their decision to leave the bank.<br>
- __Gender__ : it’s interesting to explore whether gender plays a role in a customer leaving the bank.<br>
- __Age__ : this is certainly relevant, since older customers are less likely to leave their bank than younger ones.<br>
- __Tenure__ : refers to the number of years that the customer has been a client of the bank. Normally, older clients are more loyal and less likely to leave a bank.<br>
- __Balance__ : also a very good indicator of customer churn, as people with a higher balance in their accounts are less likely to leave the bank compared to those with lower balances.<br>
- __NumOfProducts__ : refers to the number of products that a customer has purchased through the bank.<br>
- __HasCrCard__ : denotes whether or not a customer has a credit card. This column is also relevant, since people with a credit card are less likely to leave the bank.<br>
- __IsActiveMember__ : active customers are less likely to leave the bank.<br>
- __EstimatedSalary__ : as with balance, people with lower salaries are more likely to leave the bank compared to those with higher salaries.<br>
- __Exited__ : whether or not the customer left the bank.<br>
- __Complain__ : customer has complaint or not.<br>
- __Satisfaction Score__ : Score provided by the customer for their complaint resolution.<br>
- __Card Type__ : type of card hold by the customer.<br>
- __Points Earned__ : the points earned by the customer for using credit card.<br>

# Churn Rate

Out of a total of 10,000 customers, approximately 20.38% of them have decided to leave the bank. This churn rate is relatively high, implying that, on average, one out of every five customers chooses to stop doing business with the bank. This underscores the significance of understanding the factors that impact customer churn rate.

# Gender

In the dataset, the number of males exceeds the number of females. To ensure there is no bias, we calculate the customer churn rate based on the total count of each gender. The two pies above illustrate the customer churn rates for females and males, which are 25.07% and 16.47%, respectively. These results indicate that the churn rate is higher among female customers compared to male customers.

# Age

Based on the age distribution, the majority of customers fall within the 30 to 45 years old range. To investigate further, we created box plots of age categorized by churn status, and the results demonstrate that older customers are more likely to leave the bank compared to younger ones. A  detailed analysis of churn among different age groups is presented in the next section.

# Age Group

The distribution of age groups is consistent with the age distribution, with the majority falling into the 31 to 40 age group, accounting for 44.51% of the customers, followed by the 41 to 50 age group at 23.2%. Upon examining the bar chart of age groups and churn, we observe that the churn rate for the 41 to 50 age group increases significantly to 33.97%, compared to 12.11% for the 31 to 40 age group. The churn rate is even more severe for the 51 to 60 age group, surpassing half of the customers at 56.21%. This highlights a critical churn issue among customers in the 51 to 60 age range.

# Geography

Let's now shift our focus to the geography aspect. The distribution graphs indicate that the number of customers residing in France is twice that of customers living in Germany and Spain. To ensure an unbiased analysis of churn rates among different regions, we compute the customer churn rate for each specific region. The results reveal that the customer churn rate in Germany is twice as high as that in both France and Spain.

# Complain

From the distribution graph on the left, it is evident that only around 2000 customers have lodged complaints. We proceeded to conduct a detailed analysis of the churn behavior for customers who have and have not lodged complaints. Surprisingly, we found that customers who have never complained do not tend to leave the bank. On the contrary, customers who have filed complaints are more likely to churn. This observation suggests that the bank might face challenges in resolving customer complaints adequately or that such complaints adversely impact the overall user experience, leading customers to decide to leave the bank.

The upcoming section will delve further into the relationship between customer satisfaction and the churn rate, providing additional insights into this important aspect.

# Satisfaction Score

The counts of each satisfaction score are evenly distributed across all satisfaction levels. Analyzing the percentage of the total count of each satisfaction score in relation to the churn rate, we observe that the churn rate is generally consistent across all satisfaction scores. However, there is a slight variation in the churn rate for customers who rated their complaint resolution as 2, with a slightly higher rate of 21.8%. This indicates that if the bank is unable to resolve customer complaints adequately, and customers rate their complaint resolution as 2, they are more likely to leave the bank compared to customers who rate the complaint resolution differently.

This finding underscores the importance of addressing customer complaints promptly and effectively to improve overall customer satisfaction and reduce churn rates.

# Tenure

With the exception of customers who are either new (0 years) or have been loyal (10 years) to the bank, the count of tenure is evenly distributed among different levels of tenure. Among all the tenure levels, new customers (0 years) have the highest churn rate, accounting for 23%, followed closely by customers with 1 year of tenure at 22.42%. After the initial period, the churn rates vary between 17.22% to 21.75% for different tenure levels.

This observation suggests that the early stages of the customer-bank relationship are critical, and efforts to improve customer retention should be particularly focused on new customers. Understanding the reasons behind the high churn rate during the initial period can help the bank implement strategies to enhance customer satisfaction and loyalty during these crucial early stages of the customer's tenure.

# Credit Card Type

Next, we will explore the relationship between credit card type and churn rate. The customers with different credit card types are evenly distributed, and the bar chart at the bottom indicates that customers with Diamond credit cards have the highest churn rate at 21.78%, while those with Gold credit cards have the lowest churn rate at 19.26%.

To address the higher churn rate among customers with Diamond credit cards, the bank should conduct a detailed analysis of the benefits offered to these customers and examine any complaints received from them. Understanding the specific pain points and concerns of Diamond cardholders can help the bank implement targeted strategies to reduce the churn rate and enhance customer retention among this particular segment. Similarly, it may be beneficial to identify the factors that contribute to the lower churn rate among Gold cardholders to implement successful retention practices across other customer segments.

# Number of Products

The first bar chart illustrates that over 90% of the customers only have 1 to 2 products associated with the bank. The second bar chart reveals a significant difference in churn rates between customers with 1 product and those with 2 products. Customers with only 1 product have a churn rate that is nearly four times higher than those with 2 products.

Regarding customers with 3 and 4 products, it is important to note that the churn rates may not be entirely accurate due to the limited count in the sample. As a result, the focus should primarily be on customers with 1 and 2 products. The data suggests that encouraging customers to have one more product associated with the bank could potentially reduce their churn rate.

This insight highlights the significance of cross-selling and upselling strategies to increase customer product adoption and loyalty. By offering additional products that align with the customers' needs and preferences, the bank can enhance customer engagement and satisfaction, leading to a decreased likelihood of churn.

# Having Credit Card & Active Member

The churn rate for customers with or without a credit card is comparable, indicating that the presence of a credit card is not a significant factor contributing to customer churn.

However, when examining the relationship between being an active member and churn rate, a noticeable trend emerges. The churn rate drops significantly from 26.87% for non-active members to 14.27% for active members. This provides clear evidence of the crucial role of customer engagement in mitigating churn. Actively engaged members, who regularly interact with the bank and utilize its services, display higher loyalty and are less likely to discontinue their relationship with the institution.

These findings underscore the importance of fostering strong customer engagement and providing valuable services to enhance customer satisfaction and retention rates. Banks should prioritize efforts to engage customers and meet their needs to build lasting and mutually beneficial relationships.

# Salary & Credit Score

The distribution of salary based on churn reveals that, on average, the salary of customers who churned is slightly higher than that of customers who chose to stay with the bank. Additionally, the box plot analysis of credit scores and churn provides further evidence that customers with lower credit scores are more prone to leaving the bank.

These observations suggest that salary and credit score could be important factors influencing customer churn. Customers with higher salaries may have different financial needs or expectations, while those with lower credit scores might face financial challenges that affect their decision to continue their banking relationship.

# Balance & Points Earned

Based on the analysis, it appears that for customers who churned, the balance of their bank accounts is generally higher than that of customers who decided to stay. This finding aligns with our observations regarding the relationship between salary and churn rate. Customers with higher balances might have different financial goals or expectations, which could influence their decision to leave the bank.

On the other hand, it seems that the points earned by customers do not significantly affect their decision to churn. Points earned may not be a significant driver in the customer's decision-making process regarding whether to stay or leave the bank.
